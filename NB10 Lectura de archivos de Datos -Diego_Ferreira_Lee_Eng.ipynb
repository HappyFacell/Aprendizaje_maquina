# 10. Lectura de archivos de Datos

En Python es posible leer los archivos más comunes para guardar datos. Este documento explica algunos ejemplos de código básico para lectura de datos.

## 10.1 Lectura de archivos de texto simples

Del siguiente link: https://archive.ics.uci.edu/ml/datasets/iris descarga la base de datos llamada "iris.data" y guardala en el mismo folder donde se encuentre este notebook de Jupyter.

Para leer archivos de texto utilizamos la instrucción $open()$:

    file = open('file_name.ext', 'r')
    
Para imprimir el contenido del archivo utilizaremos la siguiente instrucción:

    print(file.read())
    
### 1. Utiliza las instrucciones anteriores para leer e imprimir el archivo de ejemplo 'iris.data':

```python
file = open('iris.data', 'r')

print(file.read())
```

## 10.2 Lectura de archivos tipo CSV

Cuando el archivo tiene un formato en sus datos y estos se encuentran separadaos por algún caracter en específico se utiliza la librería Pandas con la siguiente instrucción:

    df = pd.read_csv('file_name.ext', sep = ',')    # lee archivos donde la primera fila es el encabezado
    
    df = pd.read_csv('file_name.ext', sep = ',', header = None)   # lee archivos que no tengan encabezados
    
### 2. Cargar en la variable df_iris el dataset del archivo 'iris.data', abrir el archivo para observar si tiene encabezados. Posteriormente imprimir sólo los primeros 5 registros.

```python
import pandas as pd

df = pd.read_csv('iris.data', sep=',', header=None)

print(df)
```

## 10.3 Lectura de archivos de Excel

Los archivos de Excel requieren más parámetros para ser leidos, por ejemplo la hoja en que se encuentran los datos o si tiene encabezados.
    
    df = pd.read_excel('file_name.xlsx',sheet_name=0, header=None) #sin encabezados
    
    df = pd.read_excel('file_name.xlsx',sheet_name=0, header=0) #con encabezados
    
### 3. Cargar un archivo de excel de formato en base de datos con extensión '*.xlsx' y que tenga encabezados, imprimir los primeros 5 registros.

```python
#With header
import pandas as pd

df = pd.read_excel('Iris_data.xlsx', sheet_name = 0, header = None)
print(df)
```
